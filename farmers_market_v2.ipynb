```python
import pandas as pd
from config import username, password
import sqlalchemy as sa
```

```python
#connected to base
engine = sa.create_engine(f'postgresql://({username}):{password}@localhost:5438/postgres')
```

```python
#read farmers_markets csv 8804 rows × 59 columns
farmer_market = pd.read_csv('resources/farmers_markets_from_usda.csv')
farmer_market.head()
```

```python
#read svi_county .csv 3143 rows × 158 columns
svi_county = pd.read_csv('resources/SVI2020_US_COUNTY.csv')
svi_county
```

```python
#created svicounty_df  with selected column fields 3143 rows × 14 columns
svicounty_df = svi_county[['STATE','COUNTY','E_TOTPOP','E_POV150','M_POV150','E_UNEMP','M_UNEMP','E_HBURD','M_HBURD','E_NOHSDP','M_NOHSDP','E_UNINSUR','M_UNINSUR','E_MINRTY','M_MINRTY','EP_MINRTY','MP_MINRTY','EPL_MINRTY']]
svicounty_df.head()
```

```python
#created farmer_market_df  with selected column fields 8804 rows × 10 columns
farmer_market_df = farmer_market[['FMID','MarketName','street','city','County','State','zip','x','y']]
farmer_market_df.head()
```

```python
#remove any null values
clean_farm_market=farmer_market_df.dropna()
clean_farm_market
```

```python
#remove any null values
clean_svicounty_df=svicounty_df.dropna()
clean_svicounty_df
```

```python
clean_farm_market.loc[(clean_farm_market['County'] == 'Northhampton') & (clean_farm_market['State'] == 'Pennsylvania'), 'County'] = 'Northampton'
clean_farm_market.loc[(clean_farm_market['County'] == 'Placer') & (clean_farm_market['State'] == 'Indiana'), 'County'] = 'Dekalb'
clean_farm_market.loc[(clean_farm_market['County'] == 'Shasta') & (clean_farm_market['State'] == 'Arkansas'), 'County'] = 'Benton'
clean_farm_market.loc[(clean_farm_market['County'] == 'Anderoscoggin') & (clean_farm_market['State'] == 'Maine'), 'County'] = 'Androscoggin'
```

```python
clean_svicounty_df['county_lower']=clean_svicounty_df['COUNTY'].str.lower()
clean_farm_market['county_lower']=clean_farm_market['County'].str.lower()
```

```python
#created clean_svicounty_dff and created merge
clean_svicounty_dff = pd.merge(clean_farm_market, clean_svicounty_df, how="outer", right_on=['county_lower','STATE'], left_on=['county_lower','State'])
clean_svicounty_dff
```

```python
clean_svicounty_dff = clean_svicounty_dff.rename(columns={
    'street':'street_farm',
    'city':'city_farm',
    'County':'county_farm',
    'State':'state_farm',
    'zip':'zip_farm',
    'x':'x_farm',
    'y':'y_farm',
    'COUNTY':'county_svi',
    'STATE':'state_svi'
})

# for error checking - delete before final push
farm_nan_df = clean_svicounty_dff.loc[(clean_svicounty_dff['county_svi'].isnull())]
farm_nan_df = farm_nan_df[['county_farm','state_farm','county_svi','state_svi','county_lower']]
farm_nan_df
```

```python
final_svicounty_dff = clean_svicounty_dff.drop(columns=['county_lower'])
final_svicounty_dff
```

```python
final_svicounty_dff.to_csv('resources/farmers_market_svi.csv')
```
